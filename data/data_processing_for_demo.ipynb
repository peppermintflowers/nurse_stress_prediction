```python
import pandas as pd
```

```python
# read complete data file
df=pd.read_csv("merged_data.csv")
print(df["id"].value_counts())
```

```python
# drop label column
cols = ["X","Y","Z","EDA","HR","TEMP","id","datetime"]
df = df[cols] 
# separate df for separate ids
df_BG = df[df["id"] == "BG"]
df_6D = df[df["id"] == "6D"]
```

```python
CSV_BG = "BG.csv"
CSV_6D = "6D.csv"

# Save files
df_BG.to_csv(CSV_BG, index=False)
df_6D.to_csv(CSV_6D, index=False)
```
